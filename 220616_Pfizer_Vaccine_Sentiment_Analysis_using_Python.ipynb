```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/amankharwal/Website-data/master/vaccination_tweets.csv')
```

```python
import nltk
import re
```

```python
nltk.download('stopwords')
stemmer = nltk.SnowballStemmer('english')
```

```python
from nltk.corpus import stopwords
import string
```

```python
stopword = set(stopwords.words('english'))
```

```python
def clean(text):
  text = str(text).lower()
  text = re.sub('\[.*?\]', '', text)
  text = re.sub('https?://\S+|www\.\S+', '', text)
  text = re.sub('<.*?>+', '', text)
  text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
  text = re.sub('\n', '', text)
  text = re.sub('\w*\d\w*', '', text)
  text = [word for word in text.split(' ') if word not in stopword]
  text = ' '.join(text)
  text = [stemmer.stem(word) for word in text.split(' ')]
  text = ' '.join(text)
  return text
```

```python
data['text'] = data['text'].apply(clean)
```

```python
text = ' '.join(i for i in data.text)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color='white').generate(text)
plt.figure(figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show();
```

```python
text = ' '.join(i for i in data.hashtags)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color='white').generate(text)
plt.figure(figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show();
```

```python
nltk.download('vader_lexicon')
sentiments = SentimentIntensityAnalyzer()
data['Positive'] = [sentiments.polarity_scores(i)['pos'] for i in data['text']]
data['Negative'] = [sentiments.polarity_scores(i)['neg'] for i in data['text']]
data['Neutral'] = [sentiments.polarity_scores(i)['neu'] for i in data['text']]
data['Compound'] = [sentiments.polarity_scores(i)['compound'] for i in data['text']]
data2 = data[['text', 'Positive', 'Negative', 'Neutral', 'Compound']]
```

```python
x = sum(data['Positive'])
y = sum(data['Negative'])
z = sum(data['Neutral'])
c = sum(data['Compound'])
```

```python
def sentiment_score(a, b, c):
  if (a>b) and (b>c):
    print('Positive')
  elif (b>a) and (b>c):
    print('Negative')
  else:
    print('Neutral')
```

```python
sentiment_score(x,y,z)
```
