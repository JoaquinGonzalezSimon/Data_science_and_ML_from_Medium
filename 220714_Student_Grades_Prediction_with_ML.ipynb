```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.utils import shuffle
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/amankharwal/Website-data/master/student-mat.csv')
```

```python
for i in data.columns:
  print(i, data[str(i)].sort_values().unique())
  print('******'*5)
```

```python
data = data[["G1", "G2", "G3", "studytime", "failures", "absences"]]
predict = 'G3'
```

```python
x = np.array(data.drop([predict], axis=1))
y = np.array(data[predict])
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2)
xtrain.shape, xtest.shape, ytrain.shape, ytest.shape
```

```python
model = LinearRegression()
```

```python
model.fit(xtrain, ytrain)
ypred = model.predict(xtest)
```

```python
accuracy = model.score(xtest, ytest)
```

```python
accuracy
```

```python
for i in range(len(ypred)):
  print(ypred[i], xtest[i], [ytest[i]])
```

```python
print('MAE: ', metrics.mean_absolute_error(ytest, ypred))
print('MAPE: ', metrics.mean_absolute_percentage_error(ytest, ypred))
print('MSE: ', metrics.mean_squared_error(ytest, ypred, squared=False))
print('RMSE: ', metrics.mean_squared_error(ytest, ypred, squared=True))
```
