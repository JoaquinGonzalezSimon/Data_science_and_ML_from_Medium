```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.utils import shuffle
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/amankharwal/Website-data/master/student-mat.csv')
```

```python
data.head()
```

```python
pd.set_option('display.max_columns', None)
```

```python
data.head()
```

```python
data.info()
```

```python
data = data[['G1', 'G2', 'G3', 'studytime', 'failures', 'absences']]
data.head()
```

```python
predict = 'G3'
```

```python
x = np.array(data.drop([predict], 1))
y = np.array(data[predict])
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
xtrain.shape, xtest.shape, ytrain.shape, ytest.shape
```

```python
linear_regression = LinearRegression()
```

```python
linear_regression.fit(xtrain, ytrain)
```

```python
import pickle
```

```python
with open('pickle_mode', 'wb') as file:
  pickle.dump(linear_regression, file)
```

```python
with open('pickle_mode', 'rb') as file:
  new_model = pickle.load(file)
```

```python
ypred = new_model.predict(xtest)
```

```python
for i in range(len(ypred)):
  print(ypred[i], xtest[i], [ytest[i]])
```

```python
from sklearn import metrics
```

```python
print('MAE: ', metrics.mean_absolute_error(ytest, ypred))
print('MAPE: ', metrics.mean_absolute_percentage_error(ytest, ypred))
print('MSE: ', metrics.mean_squared_error(ytest, ypred, squared=False))
print('RMSE: ', metrics.mean_squared_error(ytest, ypred, squared=True))
```
