```python
import pandas as pd
import numpy as np
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
```

```python
trends = TrendReq()
```

```python
trends
```

```python
trends.build_payload(kw_list=['Machine Learning'])
data = trends.interest_by_region()
```

```python
data.head()
```

```python
data.info()
```

```python
data = data.sort_values(by='Machine Learning', ascending=False)
data.head()
```

```python
data.reset_index(drop=False, inplace=True)
data.head()
```

```python
data_graph = data.head(20)
```

```python
data_graph.plot(x='geoName', y='Machine Learning', figsize=(15,10), kind='bar')
plt.style.use('fivethirtyeight')
plt.show()
```

```python
data_2 = TrendReq(hl='en-US', tz=360)
data_2.build_payload(kw_list=['Machine Learning'])
data_2
```

```python
data_2 = data_2.interest_over_time()
data_2
```

```python
fix, ax = plt.subplots(figsize=(20,10))
data_2['Machine Learning'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for machine Learning in US', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
