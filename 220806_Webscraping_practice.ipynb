```python
import re
import requests
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
```

```python
url = 'https://www.mediotiempo.com/futbol/mls/chicharito-se-disculpo-con-el-publico-por-los-recientes-malentendidos'
```

```python
response = requests.get(url)
```

```python
response
```

```python
contenido = response.content
contenido
```

```python
soup = BeautifulSoup(contenido, 'lxml')
soup
```

```python
soup.find_all('h2')
```

```python
soup.find_all('h1')
```

```python
general = soup.select("div [id='content-body']")
general
```

```python
general_2 = [i.text for i in general]
```

```python
general_2
```

```python
text = [i.strip() for i in general_2]
text
```

```python
re.findall('errores', text[0])
```

```python
text = text[0]
text
```

```python
re.findall('errores', text)
```

```python
text.find('errores')
```
