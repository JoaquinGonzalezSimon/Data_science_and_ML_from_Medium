```python
import pandas as pd
import numpy as np
import plotly.express as px
from textblob import TextBlob
```

```python
data = pd.read_csv('/content/drive/MyDrive/Ironhack/DBs/netflix_titles.csv')
```

```python
z = data.groupby('rating').size().reset_index(name='counts')
pieChart = px.pie(z, values='counts', names='rating', title='Distribution of content ratings on Netflix',
                  color_discrete_sequence=px.colors.qualitative.Set3)
pieChart.show()
```

```python
data['director'] = data['director'].fillna('No Director Specified')
filtered_directors = pd.DataFrame()
filtered_directors = data['director'].str.split(', ', expand=True).stack()
```

```python
filtered_directors = filtered_directors.to_frame()
```

```python
filtered_directors.columns=['Director']
filtered_directors
```

```python
 directors = filtered_directors.groupby('Director').size().reset_index(name='Total content')
 directors
```

```python
directors = directors[directors.Director != 'No Director Specified']
```

```python
directors = directors.sort_values(by=['Total content'], ascending=False)
directors
```

```python
directorsTop5 = directors.head()
```

```python
directorsTop5 = directorsTop5.sort_values('Total content', ascending=True)
```

```python
fig1 = px.bar(directorsTop5, x='Total content', y='Director', title='Top 5 Directors in Netflix')
fig1.show()
```

```python
data['cast'] = data['cast'].fillna('No Cast Specified')
data['cast'].value_counts()
```

```python
filtered_cast = pd.DataFrame(data['cast'].str.split(',', expand=True).stack())
filtered_cast
```

```python
filtered_cast.columns=['Actors']
filtered_cast.head()
```

```python
actors = filtered_cast.groupby('Actors').size().reset_index(name='Total Content').sort_values(by='Total Content', ascending=False)
actors
```

```python
actors = actors[actors.Actors != 'No Cast Specified']
actors
```

```python
fig2 = px.bar(actors.head().sort_values(by='Total Content', ascending=True), x='Total Content', y='Actors', title='Top 5 Actors on Netflix')
fig2.show()
```

```python
df1 = data[['type', 'release_year']]
df1
```

```python
df2 = df1.groupby(['release_year', 'type']).size().reset_index(name='Total Contents')
df2
```

```python
df3 = df2[df2.release_year >= 1990]
df3
```

```python
fig3 = px.line(df3, x='release_year', y='Total Contents', color='type', title='Trend of content produced over the years on Netflix')
fig3.show()
```

```python
df4 = data[['release_year', 'description']]
```

```python
for index, row in df4.iterrows():
  z = row['description']
  testimonials = TextBlob(z)
  p = testimonials.sentiment.polarity
  if p == 0:
    sent = 'Neutral'
  elif p > 0:
    sent = 'Positive'
  else:
    sent = 'Negative'
  df4.loc[[index,2], 'Sentiment'] = sent
```

```python
df4.head()
```

```python
df5 = df4.groupby(['release_year', 'Sentiment']).size().reset_index(name='Total contents')
df5
```

```python
df5 = df5[df5.release_year >= 1990]
df5
```

```python
fig4 = px.line(df5, x='release_year', y='Total contents', color='Sentiment', title='Sentiment of content on Netflix')
fig4.show()
```
