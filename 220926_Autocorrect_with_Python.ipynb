```python
import pandas as pd
import numpy as np
import textdistance
import re
from collections import Counter
import requests
import csv
```

```python
url = 'https://raw.githubusercontent.com/amankharwal/Website-data/master/book.txt'
response = requests.get(url)
response.raise_for_status()
response.text
```

```python
words = []
file_name_data = response.text
file_name_data = file_name_data.lower()
words = re.findall('\w+', file_name_data)
```

```python
print(words)
```

```python
v = set(words)
print(f'First 10 words: \n{words[0:10]}')
print(f'There are {len(v)} unique words in the vocabulary')
```

```python
word_freq_dict = {}
word_freq_dict = Counter(words)
word_freq_dict.most_common(10)
```

```python
probs = {}
total = sum(word_freq_dict.values())
```

```python
for k in word_freq_dict.keys():
  probs[k] = word_freq_dict[k]/total
```

```python
def my_autocorrect(input_word):
    input_word = input_word.lower()
    if input_word in v:
        return('Your word seems to be correct')
    else:
        similarities = [1-(textdistance.Jaccard(qval=2).distance(v, input_word)) for v in word_freq_dict.keys()]
        df = pd.DataFrame.from_dict(probs, orient='index').reset_index()
        df = df.rename(columns={'Index':'Word', 0:'Prob'})
        df['Similarity'] = similarities
        output = df.sort_values(['Similarity', 'Prob'], ascending=False).head()
        return output
```

```python
my_autocorrect('neverteless')
```

```python
my_autocorrect('lenght')
```
