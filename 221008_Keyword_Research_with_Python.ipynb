```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
```

```python
search = 'Hurricane'
```

```python
trends = TrendReq()
trends.build_payload(kw_list=[search])
data = trends.interest_by_region()
df = data[data[search] != 0].sort_values(by=search, ascending=False)
df
```

```python
df.reset_index().plot(x='geoName', y=search, figsize=(15,10), kind='bar')
plt.show()
```

```python
trends.trending_searches(pn='mexico')
```

```python
keyword = trends.suggestions(keyword='trabajos')
keywords = pd.DataFrame(keyword)
keywords
```
