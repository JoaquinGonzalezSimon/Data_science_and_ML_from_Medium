```python
from google.colab import files
import io
import pandas as pd

from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LogisticRegressionCV
from sklearn.model_selection import cross_val_score
```

```python
file_train = files.upload()
```

```python
file_test = files.upload()
```

```python
train = pd.read_csv(io.BytesIO(file_train['train.csv']))
test = pd.read_csv(io.BytesIO(file_test['test.csv']))
```

```python
impute_value = train.Age.median()
impute_value
```

```python
train['Age'] = train['Age'].fillna(impute_value)
test['Age'] = test['Age'].fillna(impute_value)
```

```python
train['IsFemale'] = (train['Sex'] == 'female').astype(int)
test['IsFemale'] = (test['Sex'] == 'female').astype(int)
```

```python
predictors = ['Pclass', 'IsFemale', 'Age']
xtrain = train[predictors].values
xtest = test[predictors].values
ytrain = train['Survived'].values
```

```python
model = LogisticRegression()
model.fit(xtrain, ytrain)
model.get_params()
```

```python
ypred = model.predict(xtest)
ypred[:5]
```

```python
model_cv = LogisticRegressionCV(Cs=10)
model_cv.fit(xtrain, ytrain)
```

```python
model = LogisticRegression(C=10)
scores = cross_val_score(model, xtrain, ytrain, cv=4)
scores
```
