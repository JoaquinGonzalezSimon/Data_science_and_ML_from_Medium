```python
from google.colab import files
import io

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
upload = files.upload()
```

```python
data = pd.read_csv(io.BytesIO(upload['ntrarogyaseva.csv']))
data.info()
```

```python
data.head()
```

```python
data.describe()
```

```python
data.columns
```

```python
for i in data.columns:
  print(i)
  print(data[i].value_counts())
  print('%%%%%%%%%%%'*10)
```

```python
mappings = {'MALE':'Male', 'Male(Child)': 'Boy', 'FEMALE':'Female', 'Female(Child)':'Girl'}
```

```python
data['SEX'] = data['SEX'].replace(mappings)
data['SEX'].value_counts()
```

```python
data.SEX.value_counts().plot.bar()
```

```python
print('Mean age: {}'.format(data['AGE'].mean()))
print('Median age: {}'.format(data['AGE'].median()))
print('Mode age: {}'.format(data['AGE'].mode()))
```

```python
sns.boxplot(data['AGE'])
```

```python
data[data['DISTRICT_NAME'] == 'Krishna'].head()
```

```python
for i in data['DISTRICT_NAME'].unique():
  print('District: {} Disease and Counts {}'.format(i, data[data['DISTRICT_NAME'] == i]['SURGERY'].value_counts().head(1)))
```

```python
for i in data['DISTRICT_NAME'].unique():
  print('District: {} Average Claim Amount: IND${}'.format(i, data[data['DISTRICT_NAME'] == i]['CLAIM_AMOUNT'].mean()))
```
