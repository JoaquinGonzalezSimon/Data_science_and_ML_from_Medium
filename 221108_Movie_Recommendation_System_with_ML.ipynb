```python
credits_cols_renamed = credits.rename(index=str, columns={'movie_id':'id'})
movies_merge = movies.merge(credits_cols_renamed, on='id')
movies_merge.head()
```

```python
data = movies_merge.copy()
data.info()
```

```python
data.columns
```

```python
data = data.drop(columns=['homepage', 'title_x', 'title_y', 'status', 'production_countries'])
data.info()
```

```python
data.drop('tagline', axis=1, inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
data.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfv = TfidfVectorizer(min_df=3, max_features=None, strip_accents='unicode', analyzer='word', token_pattern=r'\w{1,}', ngram_range=(1,3), stop_words='english')
```

```python
tfv_matrix = tfv.fit_transform(data['overview'])
print(tfv_matrix)
print(tfv_matrix.shape)
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
```

```python
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
print(sig[0])
```

```python
data.reset_index(inplace=True, drop=True)
```

```python
indices = pd.Series(data.index, index=data['original_title']).drop_duplicates()
print(indices)
print(indices['Newlyweds'])
print(sig[4794])
```

```python
print(list(enumerate(sig[indices['Newlyweds']])))
```

```python
print(sorted(list(enumerate(sig[indices['Newlyweds']])), key=lambda x: x[1], reverse=True))
```

```python
def give_recommendations(title, sig=sig):
  idx = indices[title]

  sig_scores = list(enumerate(sig[idx]))

  sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

  sig_scores = sig_scores[1:11]

  movie_indices = [i[0] for i in sig_scores]

  return data['original_title'].iloc[movie_indices]
```

```python
print(give_recommendations('Avatar'))
```
