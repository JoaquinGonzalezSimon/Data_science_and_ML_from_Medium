```python
import pandas as pd
import numpy as np

from datetime import date, timedelta
import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio
pio.templates.default = 'plotly_white'

from google.colab import files
import io
```

```python
upload = files.upload()
```

```python
control = pd.read_csv(io.BytesIO(upload['control_group.csv']), sep=';')
control.head()
```

```python
upload2 = files.upload()
```

```python
test = pd.read_csv(io.BytesIO(upload2['test_group.csv']), sep=';')
test.head()
```

```python
control.info()
```

```python
control
```

```python
test.info()
```

```python
control.columns, test.columns
```

```python
to_clean = control.columns[3:]
```

```python
for i in to_clean:
    control[i].fillna(value=control[i].mean(), inplace=True)
```

```python
control.info()
```

```python
ab_data = control.merge(test, how='outer').sort_values(['Date'])
ab_data.info()
```

```python
ab_data.head()
```

```python
ab_data.info()
```

```python
ab_data.reset_index(drop=True)
```

```python
ab_data['Campaign Name'].value_counts()
```

```python
cols = ab_data.columns
cols = [i.lower().replace(' ', '_').replace('#','no').replace('[usd]', 'usd') for i in cols]
cols
```

```python
ab_data.columns=cols
ab_data.info()
```

```python
fig = px.scatter(data_frame=ab_data,
                 x='no_of_impressions',
                 y='spend_usd',
                 size='spend_usd',
                 color='campaign_name',
                 trendline='ols')
fig.show()
```

```python
label = ['Total searches from CONTROL campaign', 'Total searches from TEST campaign']
counts = [sum(control['# of Searches']), sum(test['# of Searches'])]
colors = ['gold', 'lightgreen']
fig = go.Figure(data=[go.Pie(labels=label, values=counts)])
fig.update_layout(title_text='Control VS Test searches')
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=30, 
                  marker=dict(colors=colors,line=dict(color='black', width=3)))
fig.show()
```

```python
label = ['Website clicks from CONTROL campaign', 'Website clicks from TEST campaign']
counts = [sum(control['# of Website Clicks']), sum(test['# of Website Clicks'])]
colors = ['gold', 'lightgreen']
fig = go.Figure(data=[go.Pie(labels=label, values=counts)])
fig.update_layout(title_text='Control VS Test website clicks')
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=30, 
                  marker=dict(colors=colors,line=dict(color='black', width=3)))
fig.show()
```

```python
label = ['Content viewed from CONTROL campaign', 'Content viewed from TEST campaign']
counts = [sum(control['# of View Content']), sum(test['# of View Content'])]
colors = ['gold', 'lightgreen']
fig = go.Figure(data=[go.Pie(labels=label, values=counts)])
fig.update_layout(title_text='Control VS Test view content')
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=30, 
                  marker=dict(colors=colors,line=dict(color='black', width=3)))
fig.show()
```
