```python
import pandas as pd
import numpy as np

import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import seaborn as sns

import zipfile
import requests
import io
from google.colab import files
```

```python
url = 'https://www.kaggle.com/datasets/komalkhetlani/apple-iphone-data/download?datasetVersionNumber=1'
filename = 'apple_products.csv'
```

```python
upload = files.upload()
```

```python
#z = requests.get(url).content
f = zipfile.ZipFile(io.BytesIO(upload['iphone_sales.zip']))
data_csv = f.open(filename)
data = pd.read_csv(data_csv)
```

```python
data.head()
```

```python
data.info()
```

```python
cols = [i.lower().replace(' ','_') for i in data.columns]
data.columns = cols
data.info()
```

```python
for i in data.columns:
  print(data[i].value_counts())
  print('*************'*3)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
top_rated = data.sort_values(by='star_rating', ascending=False).head(10)
```

```python
top_rated.product_name
```

```python
iphones = top_rated.product_name.value_counts()
label = top_rated.product_name.values
values = top_rated.number_of_ratings
fig = px.bar(data, x=label, y=values, title='Ratings of highest rated iphones')
fig.update_xaxes(title='iphones')
fig.update_yaxes(title='number of ratings')
fig.show()
```

```python
counts = top_rated.number_of_reviews
fig = px.bar(data, x=label, y=counts, title='Ratings of highest rated iphones')
fig.update_xaxes(title='iphones')
fig.update_yaxes(title='number of reviews')
fig.show()
```

```python
fig = px.scatter(data, x='number_of_ratings', y='sale_price', size='discount_percentage', trendline='ols', title='Relationship between sales price and number of ratings').show()
```

```python
fig = px.scatter(data, x='number_of_ratings', y='discount_percentage', size='sale_price', trendline='ols', title='Relationship between sales price and discount_percentage').show()
```
