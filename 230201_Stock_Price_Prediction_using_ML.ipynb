```python
import pandas as pd
import numpy as np

import yfinance as yf

from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
def prepare_data(df, forecast_col, forecast_out, test_size):
  label = df[forecast_col].shift(-forecast_out)
  X = np.array(df[[forecast_col]])
  X = preprocessing.scale(X)
  X_lately = X[-forecast_out:]
  X = X[:-forecast_out]
  label.dropna(inplace=True)
  y = np.array(label)
  xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=test_size)
  response = [xtrain, xtest, ytrain, ytest, X_lately]
  return response
```

```python
data = yf.download("GOOG", interval='1d', period='5y')
data.head()
```

```python
data = data.reset_index(drop=False)
data.head()
```

```python
data['Date'] = pd.to_datetime(data['Date'].dt.strftime('%Y-%m-%d'))
```

```python
data.head()
```

```python
df = data.copy()
```

```python
forecast_col = 'Close'
forecast_out = 10
test_size = 0.2
```

```python
prepare_data(df, forecast_col, forecast_out, test_size)
```

```python
xtrain, xtest, ytrain, ytest, X_lately = prepare_data(df, forecast_col, forecast_out, test_size) 
```

```python
model = LinearRegression()
```

```python
model.fit(xtrain, ytrain)
```

```python
score = model.score(xtest, ytest)
```

```python
score
```

```python
forecast = model.predict(X_lately)
```

```python
response = {}
response['test_score'] = score
response['forecast_set'] = forecast
```
