```python
import pandas as pd
import numpy as np

import plotly.express as px

from autots import AutoTS
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/amankharwal/Website-data/master/COVID19%20data%20for%20overall%20INDIA.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data = data.drop('Date', axis=1)
data.head()
```

```python
data.info()
```

```python
px.bar(data, x='Date_YMD', y='Daily Confirmed')
```

```python
px.bar(data, x='Date_YMD', y='Daily Deceased')
```

```python
model = AutoTS(forecast_length=30, frequency='infer', ensemble='simple')
model = model.fit(data, date_col='Date_YMD', value_col='Daily Deceased', id_col=None)
pred = model.predict()
forecast = pred.forecast
forecast
```
